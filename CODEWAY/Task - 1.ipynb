  **To Do List Application**
  
A To-Do List application is a useful project that helps users manage
and organize their tasks efficiently. This project aims to create a
command-line or GUI-based application using Python, allowing users
to create, update, and track their to-do lists.

```python
#To Do List Application


import os

def display_menu():
    print("\nTo-Do List Menu:")
    print("1. View To-Do List")
    print("2. Add Task")
    print("3. Mark Task as Done")
    print("4. Exit")

def view_todo_list(todo_list):
    print("\nTo-Do List:")
    if not todo_list:
        print("No tasks.")
    else:
        for idx, task in enumerate(todo_list, start=1):
            print(f"{idx}. {task.strip()}")

def add_task(todo_list):
    task = input("\nEnter the task: ")
    todo_list.append(task)
    save_to_file(todo_list)
    print("Task added successfully.")

def mark_task_as_done(todo_list):
    view_todo_list(todo_list)
    try:
        task_index = int(input("\nEnter the number of the task to mark as done: ")) - 1
        if 0 <= task_index < len(todo_list):
            del todo_list[task_index]
            save_to_file(todo_list)
            print("Task marked as done.")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

def save_to_file(todo_list):
    with open("todo_list.txt", "w") as file:
        for task in todo_list:
            file.write(task + "\n")

def load_from_file():
    todo_list = []
    if os.path.exists("todo_list.txt"):
        with open("todo_list.txt", "r") as file:
            todo_list = [line.strip() for line in file.readlines()]
    return todo_list

def main():
    todo_list = load_from_file()

    while True:
        display_menu()
        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            view_todo_list(todo_list)
        elif choice == "2":
            add_task(todo_list)
        elif choice == "3":
            mark_task_as_done(todo_list)
        elif choice == "4":
            print("Exiting the To-Do List application. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
