  **Quiz Game**
  
Develop a quiz game that asks users multiple-choice or fill-in-the-blank questions on a
specific topic. The game should keep track of scores, provide feedback on
correct/incorrect answers, and offer a variety of questions to make it challenging and
engaging.

```python
#Quiz Game

import random

class Quiz:
    def __init__(self, topic, questions):
        self.topic = topic
        self.questions = questions
        self.score = 0

    def display_welcome_message(self):
        print(f"Welcome to the {self.topic} Quiz!")
        print("Answer the following questions and test your knowledge.")

    def present_quiz_questions(self):
        for idx, question in enumerate(self.questions, start=1):
            print(f"\nQuestion {idx}: {question['question']}")
            for option_idx, option in enumerate(question['options'], start=1):
                print(f"{option_idx}. {option}")

            user_answer = input("Your answer (enter the number): ")
            self.evaluate_user_answer(question, user_answer)

    def evaluate_user_answer(self, question, user_answer):
        correct_answer = question['correct_option']
        if user_answer.isdigit() and 1 <= int(user_answer) <= len(question['options']):
            user_answer = int(user_answer)
            if user_answer == correct_answer:
                print("Correct!\n")
                self.score += 1
            else:
                print(f"Incorrect! The correct answer is: {question['options'][correct_answer - 1]}\n")
        else:
            print("Invalid choice. Please enter a valid number.\n")

    def display_final_results(self):
        print(f"Quiz completed! Your final score is: {self.score}/{len(self.questions)}")
        if self.score == len(self.questions):
            print("Congratulations! You got all questions right.")
        else:
            print("Better luck next time!")

def play_quiz():

    quiz_questions = [
        {
            'question': 'What is the capital of France?',
            'options': ['Berlin', 'Madrid', 'Paris', 'Rome'],
            'correct_option': 3
        },
        {
            'question': 'Which planet is known as the Red Planet?',
            'options': ['Mars', 'Venus', 'Jupiter', 'Saturn'],
            'correct_option': 1
        },
        # You can add as many as questions you want
    ]

    # To shuffle the questions randomly
    random.shuffle(quiz_questions)


    quiz = Quiz(topic="General Knowledge", questions=quiz_questions)
    quiz.display_welcome_message()
    quiz.present_quiz_questions()
    quiz.display_final_results()

if __name__ == "__main__":
    play_quiz()
```
